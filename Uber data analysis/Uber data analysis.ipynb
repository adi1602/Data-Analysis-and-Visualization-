# Import The Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import calendar
```

```python
data = pd.read_csv(r'D:\Datasets\Uber.csv')
```

```python
data.head()
```

```python
data.isnull().sum() #Check for Mising Values
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data['START_DATE*'] = pd.to_datetime(data['START_DATE*'], format="%m/%d/%Y %H:%M")
data['END_DATE*'] = pd.to_datetime(data['END_DATE*'], format="%m/%d/%Y %H:%M")
```

```python
hour=[]
day=[]
dayofweek=[]
month=[]
weekday=[]
for x in data['START_DATE*']:
    hour.append(x.hour)
    day.append(x.day)
    dayofweek.append(x.dayofweek)
    month.append(x.month)
    weekday.append(calendar.day_name[dayofweek[-1]])
data['HOUR']=hour
data['DAY']=day
data['DAY_OF_WEEK']=dayofweek
data['MONTH']=month
data['WEEKDAY']=weekday
```

```python
data.head()
```

```python
sns.countplot(x='CATEGORY*',data=data) #Categories We Have
```

```python
data['MILES*'].plot.hist() #How long do people travel with Uber?
```

## What Hour Do Most People Take Uber To Their Destination?

```python
hours = data['START_DATE*'].dt.hour.value_counts()
hours.plot(kind='bar',color='brown',figsize=(10,5))
plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.title('Number of trips Vs hours')
```
